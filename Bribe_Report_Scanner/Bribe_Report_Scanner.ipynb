# Scanning of Bribe Report :

This dataset contains information about the bribes taken from public. All the information in this dataset is submitted by public of India. It contains information about departments which did the corruption, date, location, amount, views of their post and amount they were forced to give.

- **Problem Statements :-**
    - (1) Find out which cities pay maximum bribes
    - (2) Find out which states pay maximum bribes 
    - (3) Which department gets paid the maximum amount of bribe on average and overall
    - (4) Bribes paid by days of the week
    - (5) Bribes paid by days of the month
    - (6) Bribes paid by Years
    - (7) Bribes paid by Months

### Calculating Total No. of bribes and Sum of bribes w.r.t City,State and Department :

**w.r.t. Departments :**

```python
department_sum=df.groupby('Department')['Amount(INR)'].agg(['sum','mean']).sort_values(by='sum',ascending=False).head(30)
```

```python
department_sum.head()
```

### Observation :
- (1) The departments mentioned as **'Others'** collect bribe maximum no. of time followed by **'Police Verification for Passport'** , **'Traffic Violation'** etc. as depicted in the graph as well as in table.
